# Moon datasetの導入

```python
from matplotlib import pyplot as plt
from sklearn import datasets
X, y = datasets.make_moons(60, noise=0.2, random_state=2020)

a0, b0 = X[y==0, 0], X[y==0, 1]
a1, b1 = X[y==1, 0], X[y==1, 1]

plt.figure(figsize=(8, 7))
plt.scatter(a0, b0, marker='o', c='blue', s=40, label='y=-1')
plt.scatter(a1, b1, marker='o', c='red', s=40, label='y=1')
plt.legend()
plt.xlabel('a')
plt.ylabel('b')
plt.show()
```
