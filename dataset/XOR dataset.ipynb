# Xor datasetの導入

```python
import random
import matplotlib.pyplot as plt

X0 = []
X1 = []
y = []
for i in range(100):
    x0 = random.uniform(-1,1)
    x1 = random.uniform(-1,1)
    X0.append(x0)
    X1.append(x1)
  
    if (x0>0 and x1<0) or (x0<0 and x1>0):
        y.append(1)
    else:
        y.append(0) 


a0 = [x0 for i, x0 in enumerate(X0) if i in [i for i, y in enumerate(y) if y==0]]
b0 = [x1 for i, x1 in enumerate(X1) if i in [i for i, y in enumerate(y) if y==0]]

a1 = [x0 for i, x0 in enumerate(X0) if i in [i for i, y in enumerate(y) if y==1]]
b1 = [x1 for i, x1 in enumerate(X1) if i in [i for i, y in enumerate(y) if y==1]]


plt.figure(figsize=(8,7))
plt.scatter(a0, b0, marker='o', c='blue', s=40, label='y=-1')
plt.scatter(a1, b1, marker='o', c='red', s=40, label='y=1')
plt.legend()
plt.xlabel('a')
plt.ylabel('b')
plt.show()
```
